```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
p = np.array([5, 1, 4, 6, 2, 4])
q1 = p
q2 = np.array([3, 7, 1, 4, 1, 6])
q3 = np.array([2, 5, 7, 2, 1, 5])
```

Calculating cross entropy loss function H(p, q) for p and q1, q2 and q3

```python
-sum(p*np.log(q1))
```

```python
-sum(p*np.log(q2))
```

```python
-sum(p*np.log(q3))
```

H(p, q) = -sum(p*np.log(q)) is smallest when p = q1

Turns out that the set of numbers that minimizes this is exactly q = p (Can be proven mathematically)

```python
p = np.zeros(10, dtype = int); p[4] = 1
q = np.random.rand(10)
q = q/sum(q)
```

```python
q
```

In the case where it is known which class an image belongs to:

```python
H = -np.log(q[p>0])
H
```

```python
q[4] = 20
q = q/sum(q) #Renormalize
H = -np.log(q[p>0])
H
```
