Torch Starts to differ from Numpy in terms of automatically computing gradient of operations

```python
x = torch.tensor([[5., 8.], [4., 6.]], requires_grad = True)
x
```

```python
y = x.pow(3).sum()
y
```

Compute the gradient:

```python
y.backward() #Compute the gradient
x.grad #Print the gradient (Everything that has happened to x)
```

Double check using the analytical formula (3x^2):

```python
3*x**2
```

Pytorch is also faster for matrix multiplication than numpy

Using pytorch:

```python
A = torch.randn(1000, 1000)
B = torch.randn(1000, 1000)
```

```python
t1 = time.perf_counter()
torch.matmul(A, B)
t2 = time.perf_counter()
print(t2 - t1)
```

Using numpy:

```python
A = np.random.randn(1000, 1000)
B = np.random.randn(1000, 1000)
```

```python
t1 = time.perf_counter()
A@B
t2 = time.perf_counter()
print(t2 - t1)
```
